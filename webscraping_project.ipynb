```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import numpy as np
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_by_revenue'
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text, 'html')
print(soup.prettify())
```

```python
tab = soup.find('table')
```

```python
word_title = tab.find_all('th')[:10]
```

```python
world_table_title = [title.text.strip() for title in word_title] 
print(world_table_title)
```

```python
world_table_title[6]= world_table_title[6].replace('Headquarters[note 1]','Headquarters')
```

```python
df = pd.DataFrame(columns = world_table_title)
df
```

```python
column_data = tab.find_all('tr')
```

```python
mydata = []
for row in column_data:
    row_data = row.find_all('td')
    ind_row_data = [data.text.strip() for data in row_data]
    mydata.append(ind_row_data)

del mydata[:2]
for i in range(len(mydata)):
    mydata[i].insert(0,i+1)

mydata
for data in mydata:
    length = len(df)
    df.loc[length] = data

del df['State-owned'] 
del df['Ref.']
```
